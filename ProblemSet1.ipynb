```python
# QUESTION 3

using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace DataPro
{
    internal class AreaOfTriangle
    {
        public void triangleArea(float a, float b, float c)
        {
            float s = (a + b + c) / 2;
            Console.WriteLine("the value of s is : " + s);
            double Area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            Console.WriteLine("the area is : " + Area);
        }


        public static void Main(string[] args)
        {
            AreaOfTriangle a = new AreaOfTriangle();
            a.triangleArea(2,2,2);
        }
    }
}
```

```python
# QUESTION 4

using System;
public class Program
{
    public static void Main(string[] args)
    {
        int[] arr1 = new int[5];
        arr1[0] = 25;
        arr1[1] = 47;
        arr1[2] = 42;
        arr1[3] = 56;
        arr1[4] = 32;
       
        int[] arr2 = new int[5];
        int[] arr3 = new int[5];
        int i, j = 0, k = 0;
        for (i = 0; i < 5; i++)
        {
            if (arr1[i] % 2 == 0)
            {
                arr2[j] = arr1[i];
                j++;
            }
            else
            {
                arr3[k] = arr1[i];
                k++;
            }
        }
        Console.WriteLine("Even numbers...");
        for (i = 0; i < j; i++)
        {
            Console.WriteLine(arr2[i]);
        }
        Console.WriteLine("Odd numbers...");
        for (i = 0; i < k; i++)
        {
            Console.WriteLine(arr3[i]);

        }
    }
}
```

```python
# QUESTION 5a
using System;
class Rectangle
{
    static bool inside(int x1, int y1, int x2, int y2, int x, int y)
    {
        if (x > x1 && x < x2 && y > y1 && y < y2)
            return true; return false;
    }
    public  void RectDimen(int x, int y, int x1,int y1, int x2, int y2)
    {
        
        if (inside(x1, y1, x2, y2, x, y))
            Console.Write("True");
        else
            Console.Write("False");
    }
    public static void Main(string[] args)
    {
        Rectangle rect = new Rectangle();
        rect.RectDimen(1, 1, 0, 0, 2, 3);
        Console.WriteLine("     ");
        Rectangle rect1 = new Rectangle();
        rect1.RectDimen(-1, -1, 0, 0, 2, 3);
    }
}


```
